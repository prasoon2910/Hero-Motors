## Total Current Assets

```python
# Filter the data for Current Assets
Inventories = bs[['Year', 'Inventories ']]
Current_Investments = bs[['Year', 'Current Investments ']]
Trade_Receivables = bs[['Year', 'Trade Receivables ']]
Cash_and_Cash_Equivalents = bs[['Year', 'Cash and Cash Equivalents ']]
Bank_Balances_Other_Than_Cash_and_Cash_Equivalents = bs[['Year', 'Bank Balances Other Than Cash and Cash Equivalents ']]
Loans_Short_term = bs[['Year', 'Loans - Short-term ']]
Others_Financial_Assets_Short_term = bs[['Year', 'Others Financial Assets - Short-term ']]

# Set the figure size
plt.figure(figsize=(10, 6))

# Create line charts using seaborn
sns.lineplot(x='Year', y='Inventories ', data=Inventories, color='orange')
sns.lineplot(x='Year', y='Current Investments ', data=Current_Investments, color='green')
sns.lineplot(x='Year', y='Trade Receivables ', data=Trade_Receivables, color='red')
sns.lineplot(x='Year', y='Cash and Cash Equivalents ', data=Cash_and_Cash_Equivalents, color='blue')
sns.lineplot(x='Year', y='Bank Balances Other Than Cash and Cash Equivalents ', data=Bank_Balances_Other_Than_Cash_and_Cash_Equivalents, color='purple')
sns.lineplot(x='Year', y='Loans - Short-term ', data=Loans_Short_term, color='brown')
sns.lineplot(x='Year', y='Others Financial Assets - Short-term ', data=Others_Financial_Assets_Short_term, color='gray')

# Set the title and axis labels
plt.title('Total Current Assets')
plt.xlabel('Year')
plt.ylabel('Amount')


# Combine the line charts using matplotlib
plt.legend(['Inventories', 'Current Investments', 'Trade Receivables', 'Cash and Cash Equivalents', 'Bank Balances Other Than Cash and Cash Equivalents', 'Loans - Short-term', 'Others Financial Assets - Short-term'])
plt.show()
```

**7. Total Asset Turnover Ratio:
